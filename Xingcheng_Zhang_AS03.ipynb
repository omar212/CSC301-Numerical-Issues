# when the function is  f = sin(x)

The term will be consist with (0/0!)x^0 + (1/1!)x^1 + (0/2!)x^2 + (-1/3!)x^3 + ........
so when n = 1 and n = 2 will return the same function x. Same thing will happen in other procedure.
In the following code, I decide to hide all the functions because the polynomial will be so long when 
n = 128, 256. It will be so messy. I also put a print statement which can show the output of each function
for n. Please feel free to test.

```python
import sympy as sy
import numpy as np
from sympy.functions import sin, cos, exp, log
import matplotlib.pyplot as plt
import math

x = sy.Symbol('x')
def taylor(function, a, n):
    i = 0
    polynomial = 0
    while i <= n:
        polynomial = polynomial + (function.diff(x,i).subs(x, a))/(math.factorial(i))*(x - a)**i
        i += 1
    return polynomial

def graphOfSin(f, x_lims = [-15, 15], y_lims = [-15, 15], npoints = 700):
    x1 = np.linspace(x_lims[0], x_lims[1], npoints)
    y1 = []
    n = [1, 2, 3, 4, 8, 16, 32, 64,128,256]
    for j in n:
        func = taylor(f, 0, j)
        for k in x1:
            y1.append(func.subs(x,k))
        plt.plot(x1, y1, label = 'n = '+ str(j))
        y1 = []
    plt.plot(x1, np.sin(x1), label = 'f = sin(x)')     # give a sine graph 
    fig = plt.gcf()
    fig.set_size_inches(14.5, 10.5, forward=True)
    plt.xlim(x_lims)
    plt.ylim(y_lims)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor polynomial : f = sin(x)')
    plt.show()

f = sin(x)
graphOfSin(f)


n = [1, 2, 3, 4, 8, 16, 32, 64, 128, 256]
def getFunctions(n,f):
    functionList = []
    for i in n:
        func = taylor(f,0,i)
        functionList.append(func)
    return functionList

xlist = [1/32, 1/16, 1/8, 1/4, 1/2, 1, 2, 4]
FunctionList = getFunctions(n,f)
#print(FunctionList)
def getAnswer(FunctionList):
    for thing in range(0, len(n)):
        print("===============================================================================")
        print('when n ='+str(n[thing]))
        #print('function =' +str(FunctionList[thing]))#-------------------------->turn function off
        print(".....................................................................................................................................................................................")
        for i in xlist:
            Answer = FunctionList[thing].subs(x,i)
            measure = Answer
            actual = sin(i)
            absError = measure - actual
            absError = float('%.15f' % absError)
            relativeError = (measure - actual) / actual
            relativeError = float('%.15f' % relativeError)
            Answer = float('%.20f' % Answer)
            print('if x ='+ str(i), '---',Answer,"--Actual value is: ", actual, "--absErr is: ", absError,"--relativeErr is: ",relativeError)
    return 
answerList = getAnswer(FunctionList)
```
